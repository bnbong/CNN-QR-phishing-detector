# Feature Map & Grad-CAM 시각화

CNN 모델의 중간 레이어를 시각화하여 모델이 어떤 패턴에 반응하는지 확인합니다.

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# 환경 설정
model_path = '../best_model.h5'  # change me
test_image_path = './badqr.png'  # change me
# test_image_path = '../data/images/phishing/20250131202751-0.png'  # change me
```

```python
# 모델 로드
model = tf.keras.models.load_model(model_path)
print("모델 로드 완료")
```

```python
def preprocess_image(image_path, target_size=(128, 128)):
    """이미지 전처리"""
    img = tf.keras.preprocessing.image.load_img(image_path, color_mode='grayscale', target_size=target_size)
    img_array = tf.keras.preprocessing.image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = img_array / 255.0
    return img_array
```

```python
def visualize_feature_maps(model, img):
    """CNN 모델 Feature Map 시각화 함수"""
    layer_outputs = [layer.output for layer in model.layers if 'conv2d' in layer.name]  # only con2d layer
    activation_model = tf.keras.models.Model(inputs=model.input, outputs=layer_outputs)
    
    @tf.function(reduce_retracing=True)
    def predict_activations(img):
        return activation_model(img)
    
    activations = predict_activations(img)
    
    for i, layer_activation in enumerate(activations):
        n_features = layer_activation.shape[-1]
        size = layer_activation.shape[1]
        n_cols = n_features // 8
        display_grid = np.zeros((size * n_cols, 8 * size), dtype=np.float32)  # float32
        
        for col in range(n_cols):
            for row in range(8):
                channel_image = layer_activation[0, :, :, col * 8 + row]
                channel_image -= tf.reduce_mean(channel_image)
                channel_image /= tf.math.reduce_std(channel_image)
                channel_image *= 64
                channel_image += 128
                channel_image = np.clip(channel_image, 0, 255).astype('uint8')
                display_grid[col * size : (col + 1) * size,
                            row * size : (row + 1) * size] = channel_image
        
        plt.figure(figsize=(8, 8))
        plt.title(f'Feature Maps - Layer {i+1}')
        plt.imshow(display_grid, aspect='auto', cmap='viridis')
        plt.show()
```

```python
def grad_cam(model, img, layer_name):
    """Grad-CAM 모델 결정 과정 시각화 함수"""
    grad_model = tf.keras.models.Model(
        inputs=[model.inputs],
        outputs=[model.get_layer(layer_name).output, model.output]
    )

    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img)
        loss = predictions[:, 1]  # 피싱 클래스에 대한 출력
    
    grads = tape.gradient(loss, conv_outputs)[0]
    output = conv_outputs[0]
    
    weights = tf.reduce_mean(grads, axis=(0, 1))
    cam = np.zeros(output.shape[0:2], dtype=np.float32)
    
    for index, w in enumerate(weights):
        cam += w * output[:, :, index]
    
    cam = cv2.resize(cam.numpy(), (128, 128))
    cam = np.maximum(cam, 0)
    heatmap = (cam - cam.min()) / (cam.max() - cam.min())
    
    return heatmap
```

```python
def guided_grad_cam(model, img, layer_name):
    """Guided Grad-CAM 모델 결정 과정 시각화 함수"""
    img_tensor = tf.convert_to_tensor(img, dtype=tf.float32)
    
    grad_model = tf.keras.models.Model(
        inputs=[model.inputs],
        outputs=[model.get_layer(layer_name).output, model.output]
    )
    
    with tf.GradientTape() as tape:
        tape.watch(img_tensor)
        conv_outputs, predictions = grad_model(img_tensor)
        loss = predictions[:, 1]  # 피싱 클래스에 대한 출력
    
    grads = tape.gradient(loss, conv_outputs)[0]
    output = conv_outputs[0]
    
    weights = tf.reduce_mean(grads, axis=(0, 1))
    cam = np.zeros(output.shape[0:2], dtype=np.float32)
    
    for index, w in enumerate(weights):
        cam += w * output[:, :, index]
    
    cam = cv2.resize(cam.numpy(), (128, 128))
    cam = np.maximum(cam, 0)
    heatmap = (cam - cam.min()) / (cam.max() - cam.min())
    
    # Guided Backpropagation
    guided_model = tf.keras.models.Model(
        inputs=[model.inputs],
        outputs=[model.get_layer(layer_name).output]
    )
    
    with tf.GradientTape() as tape:
        tape.watch(img_tensor)
        outputs = guided_model(img_tensor)
    
    guided_grads = tape.gradient(outputs, img_tensor)[0]
    guided_grads = np.maximum(guided_grads, 0)  # ReLU
    
    # Guided Grad-CAM
    guided_cam = heatmap[..., np.newaxis] * guided_grads
    guided_cam = cv2.resize(guided_cam, (128, 128))
    
    return guided_cam
```

```python
# Feature Map 시각화
img = preprocess_image(test_image_path)

visualize_feature_maps(model, img)
```

```python
# Grad-CAM 시각화
grad_cam = grad_cam(model, img, 'conv2d_2')
plt.figure(figsize=(8, 8))
plt.title('Grad-CAM')
plt.imshow(grad_cam, cmap='viridis')
plt.show()
```

```python
# Guided Grad-CAM 시각화
guided_cam = guided_grad_cam(model, img, 'conv2d_2')
plt.figure(figsize=(8, 8))
plt.title('Guided Grad-CAM')
plt.imshow(guided_cam, cmap='viridis')
plt.show()
```
